To characterize the motion in a flow field, the optical measuring method [Particle Image Velocimetry (PIV)] (https://de.wikipedia.org/wiki/Particle_Image_Velocimetry) is often used.
For this purpose, particles introduced into the fluid are illuminated with a laser beam that has been widened to the plane and photographed at short intervals in succession.
From the change in location of the particles and the known distance between the images, the velocity vectors can be determined.

Functions are to be written, with the help of which calculation and visualization of a velocity vector field based on two recorded images can be realized.
