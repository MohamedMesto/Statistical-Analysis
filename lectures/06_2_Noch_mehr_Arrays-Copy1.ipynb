```python
######## #    linspace

# Hier eigenen Code schreiben ...
def signal(samples_per_period = 100, n_periods=1, sigma=0.0):
    n_samples = samples_per_period*n_periods
    x = linspace(0,n_periods*2*pi,n_samples,endpoint = False)
    y = sin(x) + randn(n_samples) * sigma
    return x,y
    
```
